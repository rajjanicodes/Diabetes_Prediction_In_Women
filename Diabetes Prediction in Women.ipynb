```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score 
```

## Data collection and Analysis

PIMA Diabetes Dataset

```python
#Loading the data
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/diabetes.csv')
```

```python
data.head()
```

```python
#Checking the data shape
data.shape
```

```python
data.describe()
```

```python
data['Outcome'].value_counts()
```

```python
data.groupby('Outcome').mean()
```

```python
#separating data and labels
X = data.drop(columns='Outcome', axis=1)
Y = data['Outcome']
```

## Data Standarsization

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
stdata = scaler.transform(X)
```

```python
stdata
```

```python
X = stdata
Y = data['Outcome']
```

```python
X
```

```python
Y
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X,Y,test_size = 0.2, stratify=Y, random_state=2 )
```

```python
X.shape
```

```python
Xtrain.shape
```

```python
Xtest.shape
```

## Training the model

```python
classifier = svm.SVC(kernel='linear')
```

```python
#training the SVM
classifier.fit(Xtrain,Ytrain)
```

## Model Evaluation

Accuracy Score

```python
#accuracy score on training data
XtrainPred = classifier.predict(Xtrain)
XtrainAcc = accuracy_score(XtrainPred, Ytrain)
```

```python
print('Accuracy score of the training data : ', XtrainAcc)
```

```python
#accuracy score on TEST data
XtestPred = classifier.predict(Xtest)
XtestAcc = accuracy_score(XtestPred, Ytest)
```

```python
print('Accuracy score of the training data : ', XtestAcc)
```

##Making a Predictive System

```python
manual_input = (50,110,92,0,0,37.6,0.191,30)
dataarray = np.asarray(manual_input)
dataarrayreshaped = dataarray.reshape(1,-1)
datastd = scaler.transform(dataarrayreshaped)
#print(datastd)

prediction = classifier.predict(datastd)
#print(prediction)

if prediction[0] == 0: print('The person is NOT DIABETIC.')
else: print('The person is DIABETIC !!')
```
